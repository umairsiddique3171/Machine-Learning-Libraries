```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1. relplot (Relation Plot)

```python
# if want to plot for condition, we can do it by filtering data frame along the way 
sns.relplot(x = 'total_bill',y = 'tip',data = tips.loc[tips.sex == 'Female'])
```

```python
# if want to see the results for all condition of any particular column then we can do this
sns.relplot(x = 'total_bill',y = 'tip',data = tips , hue = 'sex',style = 'time')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips , hue = 'size')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips , hue = 'size',palette = 'Reds' )
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips , size = 'size')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips , size = 'size',sizes = (30,150))
```

```python
# if we provide any column to hue then it will change color, if we provide to size then it will change size,
# if we provide to style then it will change style.
```

```python
crash_df = sns.load_dataset('car_crashes')
```

### Distribution Plots

```python
# Distribution Plot
```

```python
sns.distplot(crash_df["not_distracted"], kde = False,bins = 25)
```

```python
# Joint Plot
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol',data = crash_df , kind = 'hex')
# here kind can also reg,kde 
```

```python
# KDE Plot
```

```python
# sns.kdeplot(x = 'alcohol',data = crash_df)
sns.kdeplot(crash_df['alcohol'])
```

```python
# Pair Plots
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.pairplot(tips_df,hue = 'sex',palette = 'Reds')
```

```python
# Rug Plot
```

```python
sns.rugplot(tips_df['tip'])
```

```python
# Styling
```

```python
sns.set_style('ticks')

plt.figure(figsize = (8,4))

#sns.set_context('poster',font_scale = 1.4)
sns.set_context('paper',font_scale = 1.4)

sns.jointplot(x = 'total_bill',y = 'tip',data = tips_df,kind = 'reg')

#sns.despine(top = True,left = True,right = True,bottom = True)
```

### Categorical Plots

```python
# Bar Plots
```

```python
sns.barplot(x = 'sex', y = 'total_bill',data = tips_df, estimator = np.std)
#estimator = np.median, np.std,np.var
```

```python
# Count Plot
```

```python
sns.countplot(x='sex',hue = 'day',data= tips_df)
```

```python
# Box Plot
```

```python
sns.boxplot(x='day',y='total_bill',data = tips_df,hue = 'sex')
```
